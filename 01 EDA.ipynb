# EDA Applications

```python
import pandas as pd
apps = pd.read_csv(
    'data/applications.csv',
    parse_dates=['orig_month'],
    thousands=','
)
apps
```

```python
apps.isna().sum()
```

```python
apps.columns = apps.columns.str.strip().str.lower()
apps.columns
```

```python

apps['income'] = apps['income'].fillna(apps['income'].median())
apps
```

```python
apps.duplicated().sum()
```

```python
apps.describe()
```

```python
import matplotlib.pyplot as plt
apps[['age','income','debt_income','bureau_score']].hist(figsize=(12,8))
plt.suptitle('Client Profile Distributions')
plt.show()
```

```python
apps.groupby('region')['income'].mean()
```

```python
apps.groupby('product')['amount'].mean()
```

```python
risk_vars = ['debt_income','utilization','prev_delin_24m','bureau_score']

apps[risk_vars].hist(figsize=(12,8))
plt.suptitle('Risk Indicators')
plt.show()
```

```python
apps[apps['debt_income'] > 1]
```

```python
apps['product'].value_counts().plot(kind='bar')
plt.title('Product Mix')
plt.show()

apps['channel'].value_counts().plot(kind='bar')
plt.title('Channel Mix')
plt.show()
```

```python
apps['orig_month'] = pd.to_datetime(apps['orig_month'])

macro = apps.groupby('orig_month')[['unemp','rate','infl']].mean()

macro.plot(figsize=(12,6))
plt.title('Macro Trends Over Time')
plt.show()
```

```python
apps.groupby('orig_month')['income'].mean().plot()
plt.title('Income Over Time')
plt.show()

apps.groupby('orig_month')['bureau_score'].mean().plot()
plt.title('Credit Quality Over Time')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(apps.select_dtypes(include='number').corr(), cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```
